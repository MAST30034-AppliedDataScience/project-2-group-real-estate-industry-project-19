Heatmap using Pearson correlation coefficients

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# File path for the merged dataset
file_path = '/home/Daniel Bi/project two/data/landing/rental_merged.csv'

# Step 1: Load the dataset
df = pd.read_csv(file_path)

# Step 2: Inspect the dataset to verify column names
print(df.columns)  # This will help you confirm the column names

# Step 3: Select relevant numeric features for correlation analysis
# Replace the old column names with the correct ones
numeric_features = ['Cost', 'Beds', 'Baths', 'Cars', 'distance_to_cbd_km', 
                    'straight_line_distance_km', 'route_distance_to_closest_train_km', 
                    'straight_line_distance_parks', 'route_distance_parks', 
                    'straight_line_distance_shopping', 'route_distance_shopping']

# Filter the DataFrame to include only these numeric features
df_numeric = df[numeric_features]

# Step 4: Compute the correlation matrix
corr_matrix = df_numeric.corr()

# Step 5: Plot the heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix: Rental Prices and Features')
plt.show()


```
